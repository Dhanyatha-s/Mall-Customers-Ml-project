```python
x= df.iloc[:,[3,4]].values
```

```python
# find the value of k k----> no of clusters
# elbow method is used to find k
# k vlaiue cannot be greater thsn sq.root of total no of points
```

```python
from sklearn.cluster import KMeans
k=range(2,15)
sse=[]
for i in k:
  model_demo=KMeans(n_clusters=i,random_state=0)
  model_demo.fit(x)
  sse.append(model_demo.inertia_)
plt.scatter(k,sse)
plt.plot(k,sse)
plt.show()
```

```python
# from the above graph ,elbow is very prominent k=5,hence we assume our k value as 5
```

```python
k=5
from sklearn.cluster import KMeans
model=KMeans(n_clusters=k,random_state=0)
model.fit(x)
```

```python
  y=model.predict(x)  
```
